#### Ahora que sabemos cómo definir transformaciones de coordenadas, las usaremos para obtener una lista de estrellas que están en GD-1. Como mencionamos antes, son muchas estrellas, así que vamos a comenzar definiendo un rectángulo que abarque una pequeña parte de GD-1. Esto es más fácil de definir en coordenadas GD-1.

#### Las siguientes variables definen los límites del rectángulo en φ1 y φ2

### Para transformar hacia y desde las coordenadas GD-1, usaremos un marco definido por Gala, que es una biblioteca afiliada a Astropy que proporciona herramientas para la dinámica galáctica.

### Gala proporciona GD1Koposov10, que es "un sistema de coordenadas esférico heliocéntrico definido por la órbita de la corriente GD-1". En este sistema de coordenadas, un eje se define a lo largo de la dirección del vapor (el eje x en la Figura 1) y un eje se define perpendicular a la dirección de la corriente (el eje y en la Figura 1). Estas se denominan coordenadas φ1 y φ2, respectivamente.

```python
def make_rectangle(x1, x2, y1, y2):
    """Return the corners of a rectangle."""
    xs = [x1, x1, x2, x2, x1]
    ys = [y1, y2, y2, y1, y1]
    return xs, ys
```

```python
def skycoord_to_string(skycoord):
    """Convert a one-dimenstional list of SkyCoord to string for Gaia's query format."""
    corners_list_str = skycoord.to_string()
    corners_single_str = ' '.join(corners_list_str)
    return corners_single_str.replace(' ', ', ')
```

```python
def make_dataframe(table):
    """Transform coordinates from ICRS to GD-1 frame.
    
    table: Astropy Table
    
    returns: Pandas DataFrame
    """
    #Create a SkyCoord object with the coordinates and proper motions
    # in the input table
    skycoord = SkyCoord(
               ra=table['ra'], 
               dec=table['dec'],
               pm_ra_cosdec=table['pmra'],
               pm_dec=table['pmdec'], 
               distance=8*u.kpc, 
               radial_velocity=0*u.km/u.s)

    # Define the GD-1 reference frame
    gd1_frame = GD1Koposov10()

    # Transform input coordinates to the GD-1 reference frame
    transformed = skycoord.transform_to(gd1_frame)

    # Correct GD-1 coordinates for solar system motion around galactic center
    skycoord_gd1 = reflex_correct(transformed)

    #Add GD-1 reference frame columns for coordinates and proper motions
    table['phi1'] = skycoord_gd1.phi1
    table['phi2'] = skycoord_gd1.phi2
    table['pm_phi1'] = skycoord_gd1.pm_phi1_cosphi2
    table['pm_phi2'] = skycoord_gd1.pm_phi2

    # Create DataFrame
    df = table.to_pandas()

    return df
```

```python
def plot_cmd(table):
    """Plot a color magnitude diagram.
    
    table: Table or DataFrame with photometry data
    """
    y = table['g_mean_psf_mag']
    x = table['g_mean_psf_mag'] - table['i_mean_psf_mag']

    plt.plot(x, y, 'ko', markersize=0.3, alpha=0.3)

    plt.xlim([0, 1.5])
    plt.ylim([22, 14])

    plt.ylabel('Magnitude (g)')
    plt.xlabel('Color (g-i)')
```
